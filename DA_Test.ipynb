```python
# Import few libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import the dataset
data = pd.read_csv("C:/Users/LG_Zarrief/Documents/DA_Test/DA_Test.csv")
```

```python
# Check first 5 rows 
data.head()
```

```python
# Adding column in dataframe
data['CR'] = (data['Orders']/data['Visits'])*100
data['ABS'] = data['Revenue in $']/data['Orders']
data['CIR'] = data['Spend in $']/data['Revenue in $']
```

```python
data
```

```python
# Visualizing Conversion Rate (CR) = Orders/Visits (%)
data.pivot(index = 'Channel name', columns = 'Week', values ='CR').plot(kind='bar', figsize=(15, 8), color=['lightcoral', 'royalblue'])

plt.xlabel('Channel Name')
plt.ylabel('Conversion Rate (%)')
plt.title('Conversion Rate by Channel')
plt.show()
```

From the figure above, it can be seen that in week 43 there are increase in conversion rate for App install network A, App install network B, Newsletters A, Web channel B	& Web channel C.

```python
# Visualizing Average Basket Size (ABS) = Revenue/Orders ($)

data.pivot(index = 'Channel name', columns = 'Week', values ='ABS').plot(kind='bar', figsize=(15, 8))

plt.xlabel('Channel Name')
plt.ylabel('Average Basket Size ($)')
plt.title('Average Basket Size by Channel')
plt.show()
```

From the figure above, it can be seen that in week 43 there are increase in Average Basket Size for App install network C & Web channel B.

```python
# Visualizing Cost to Income Ratio (CIR) = Spend/Revenue ($)
data.pivot(index = 'Channel name', columns = 'Week', values ='CIR').plot(kind='barh', figsize=(15, 8), color=['cyan', 'teal'])

plt.xlabel('Channel Name')
plt.ylabel('Cost to Income Ratio ($)')
plt.title('Cost to Income Ratio by Channel')
plt.show()
```

From the figure above, it can be seen that in week 43, almost all channels shows decrease in Cost to Income Ratio except for Web channel A & Web Channel C.
